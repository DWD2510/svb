```python
import yfinance as yf
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
def get_tesla_revenue():
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    headers = {"User-Agent": "Mozilla/5.0"}  # Tránh bị chặn request
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Kiểm tra lỗi HTTP
        soup = BeautifulSoup(response.text, "html.parser")
        
        tables = soup.find_all("table")
        for table in tables:
            if "Tesla Quarterly Revenue" in str(table):  # Nhận diện bảng bằng keyword
                revenue_df = pd.read_html(str(table))[0]
                break
        else:
            raise ValueError("Không tìm thấy bảng doanh thu Tesla")
        
        
        revenue_df.columns = ["Date", "Revenue"]  # Đặt tên cột
        revenue_df["Revenue"] = revenue_df["Revenue"].str.replace(r"[$,\s]", "", regex=True).astype(float)
        revenue_df = revenue_df.dropna(subset=["Revenue"])  
        return revenue_df.tail()  
    except Exception as e:
        print(f"Lỗi khi trích xuất dữ liệu: {e}")
        return None

tesla_revenue = get_tesla_revenue()
print(tesla_revenue)
```

```python
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
def get_gamestop_revenue():
    url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
    headers = {"User-Agent": "Mozilla/5.0"} 
    try:
       
        response = requests.get(url, headers=headers)
        response.raise_for_status()  
        soup = BeautifulSoup(response.text, "html.parser")
        
        
        tables = soup.find_all("table")
        for table in tables:
            if "GameStop Quarterly Revenue" in str(table):
                revenue_df = pd.read_html(str(table))[0]
                break
        else:
            raise ValueError("Không tìm thấy bảng doanh thu GameStop")
        
        # Chuẩn hóa dữ liệu
        revenue_df.columns = ["Date", "Revenue"]  
        revenue_df["Revenue"] = revenue_df["Revenue"].str.replace(r"[$,\s]", "", regex=True).astype(float)
        revenue_df = revenue_df.dropna(subset=["Revenue"])  
        return revenue_df.tail()  
    
    except Exception as e:
        print(f"Lỗi: {e}")
        return None


gme_revenue = get_gamestop_revenue()
print(gme_revenue)
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, revenue_data, company_name):
    """
    Creates a dual-axis graph comparing stock prices and revenue.
    
    Parameters:
        stock_data (DataFrame): Contains 'Date' and 'Close' columns
        revenue_data (DataFrame): Contains 'Date' and 'Revenue' columns
        company_name (str): Name of the company for title
    """
    plt.figure(figsize=(14, 7))
    
    # Primary axis for stock price
    plt.plot(stock_data['Date'], stock_data['Close'], 
             color='#0066CC', linewidth=2, label='Closing Price')
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Stock Price (USD)', fontsize=12, color='#0066CC')
    plt.tick_params(axis='y', labelcolor='#0066CC')
    plt.grid(True, linestyle=':', alpha=0.7)
    
    # Secondary axis for revenue
    ax2 = plt.twinx()
    ax2.plot(revenue_data['Date'], revenue_data['Revenue']/1e9,  # Convert to billions
             color='#FF6600', linestyle='--', label='Revenue')
    ax2.set_ylabel('Revenue (Billion USD)', fontsize=12, color='#FF6600')
    ax2.tick_params(axis='y', labelcolor='#FF6600')
    
    # Title and legend
    plt.title(f'{company_name} Stock Performance vs Revenue\n2010-2023', 
              fontsize=14, pad=20, weight='bold')
    lines, labels = plt.gca().get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    plt.legend(lines + lines2, labels + labels2, loc='upper left')
    
    plt.tight_layout()
    plt.show()

# Generate Tesla graph
make_graph(
    stock_data=tesla_data,
    revenue_data=tesla_revenue,
    company_name="Tesla (TSLA)"
)
```

```python
def make_gme_graph(stock_data, revenue_data):
    """
    Creates a dual-axis graph comparing GameStop stock prices and revenue.
    
    Parameters:
        stock_data (DataFrame): Contains 'Date' and 'Close' columns
        revenue_data (DataFrame): Contains 'Date' and 'Revenue' columns
    """
    plt.figure(figsize=(14, 7))
    
    # Primary axis for stock price
    plt.plot(stock_data['Date'], stock_data['Close'], 
             color='#2E8B57', linewidth=2, label='Closing Price')  # Forest Green
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Stock Price (USD)', fontsize=12, color='#2E8B57')
    plt.tick_params(axis='y', labelcolor='#2E8B57')
    plt.grid(True, linestyle=':', alpha=0.7)
    
    # Secondary axis for revenue
    ax2 = plt.twinx()
    ax2.plot(revenue_data['Date'], revenue_data['Revenue']/1e9,  # Convert to billions
             color='#8B4513', linestyle='--', label='Revenue')  # Saddle Brown
    ax2.set_ylabel('Revenue (Billion USD)', fontsize=12, color='#8B4513')
    ax2.tick_params(axis='y', labelcolor='#8B4513')
    
    # Title and legend
    plt.title('GameStop (GME) Stock Performance vs Revenue\n2002-2023', 
              fontsize=14, pad=20, weight='bold')
    lines, labels = plt.gca().get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    plt.legend(lines + lines2, labels + labels2, loc='upper left')
    
    plt.tight_layout()
    plt.show()

# Generate GameStop graph
make_gme_graph(
    stock_data=gme_data,
    revenue_data=gme_revenue
)
```
